```python
#dataset
import numpy as np
X=np.array(([2,9],[1,5],[3,6]), dtype=float)
Y=np.array(([92],[86],[89]),dtype=float)
X=X/np.amax(X,axis=0)   # take the max for each col, normalize
actual_op=Y/100
print(X)
print(actual_op)
```

```python
#activation functions
def sigmoid(x):
    return 1/(1+np.exp(-x))
def derivative_sigmoid(x):
    return x*(1-x)
```

```python
#feed forward network
#np.random.seed(1)
epoch=1000
lr=0.15
ip_neuron=2; hidden_neuron=3; op_neuron=1
wh=np.random.uniform(low=-0.05,high=0.05,
                    size=(ip_neuron,hidden_neuron)) #2X3
#print(wh)
bh=np.random.uniform(low=-0.05,high=0.05,
                    size=(1,hidden_neuron))
wo=np.random.uniform(low=-0.05,high=0.05,
                    size=(hidden_neuron,op_neuron))   #3X1
bo=np.random.uniform(low=-0.05,high=0.05,
                    size=(1,op_neuron))

print(f'wh{wh} \n bh{bh} \n wo{wo} \n bo{bo}')
```

```python
#train the ANN using backpropagation

for i in range(epoch):
    net_h=np.dot(X,wh)+bh
    sigma_h=sigmoid(net_h)
    
    net_o=np.dot(sigma_h,wo)+bo
    output=sigmoid(net_o)
    
    eo=actual_op-output
    outgrad=derivative_sigmoid(output)
    d_output=eo*outgrad    #err =oj(1-oj)(tj-oj)
    
    eh=d_output.dot(wo.T)
    hiddengrad=derivative_sigmoid(sigma_h)
    d_hidden=eh*hiddengrad
    
    wo+=sigma_h.T.dot(d_output)*lr
    wh+=X.T.dot(d_hidden)*lr
    
    sum_error=0
    for j in range(len(actual_op)):
        print(output[j],actual_op[j])
        sum_error+=((output[j]-actual_op[j])**2)
    
    sum_error/=2
    print(f'Epoch {i} error {sum_error}')
    

```
