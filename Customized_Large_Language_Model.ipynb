```python
!pip install transformers
!pip install torch
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
# Load pre-trained model and tokenizer
model_name = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
# Set the model to evaluation mode and move to GPU
model.eval()
model.to("cuda")
```

```python
def chat_with_bot(input_text):
  # Encode the user input and generate a response
  input_ids = tokenizer.encode(input_text + tokenizer.eos_token, return_tensors="pt").to("cuda")
  chat_history_ids = model.generate(input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)

  # Decode the response
  response = tokenizer.decode(chat_history_ids[:, input_ids.shape[-1]:][0], skip_special_tokens=True)
  return response

# Example conversation
user_input = "Hi, How are you?"
response = chat_with_bot(user_input)
print(f"Bot: {response}")
```

```python
import gradio as gr
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load pre-trained model and tokenizer
model_name = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Set the model to evaluation mode and move to GPU
model.eval()
model.to("cuda")

# Define the function to handle chatbot responses
def chat_with_bot(input_text, chat_history=[]):
    # Encode the user input and generate a response
    new_user_input_ids = tokenizer.encode(input_text + tokenizer.eos_token, return_tensors='pt').to("cuda")
    bot_input_ids = torch.cat([chat_history, new_user_input_ids], dim=-1) if chat_history else new_user_input_ids
    chat_history_ids = model.generate(bot_input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)

    # Get the response
    response = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)

    # Update chat history
    chat_history = torch.cat([bot_input_ids, chat_history_ids], dim=-1) if chat_history else chat_history_ids

    return response, chat_history

# Set up the Gradio interface
iface = gr.Interface(
    fn=chat_with_bot,
    inputs=[gr.Textbox(lines=2, placeholder="Type your message here..."), gr.State()],
    outputs=[gr.Textbox(), gr.State()],
    title="DialoGPT Chatbot",
    description="Chat with the DialoGPT-medium model. Type your message and get a response from the bot."
)

# Launch the Gradio interface
iface.launch()
```
